```python
Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
     MongoDB over SQL databases?
Ans:-MongoDB is a popular document-oriented NoSQL database that stores data in a JSON-like format called BSON (Binary JSON).
     It is designed to handle unstructured data and offers high scalability and performance.

    Non-relational databases, also known as NoSQL databases, are a type of database that does not use the traditional relational 
    data model used by SQL databases. Instead, they use a variety of data models such as document, key-value, column-family, 
    and graph to store and manipulate data. Non-relational databases are generally more flexible and scalable than SQL databases, 
    making them suitable for handling large volumes of unstructured or semi-structured data.
    
    
    MongoDB is preferred over SQL databases in scenarios where:

    Flexibility and scalability are key requirements: MongoDB allows you to store unstructured data without the need for predefined schemas,
    making it easy to adapt to changing requirements. It also offers horizontal scalability through sharding, which allows you to distribute 
    data across multiple nodes.

    Performance is critical: MongoDB's document-oriented data model and efficient indexing mechanisms make it faster than SQL databases for
    certain types of queries, such as those involving complex data structures.

    Real-time analytics are required: MongoDB's ability to handle real-time streaming data and its support for aggregation pipelines make 
    it well-suited for real-time analytics use cases.

    Cloud-based deployment is preferred: MongoDB's cloud-based deployment options, including its Atlas cloud service,
    make it easy to deploy and manage databases in the cloud.

    Overall, MongoDB is a powerful NoSQL database that is well-suited for handling large volumes of unstructured data, real-time analytics,
    and cloud-based deployment scenarios.
```

```python
Q2. State and Explain the features of MongoDB.
Ans:-
    1. Document-oriented data model: MongoDB stores data in flexible JSON-like documents.
    2. High performance: MongoDB is faster than traditional SQL databases for certain types of queries.
    3. Horizontal scalability: MongoDB allows you to scale horizontally using sharding.
    4. Rich query language: MongoDB supports a wide range of operations, including geospatial queries and aggregation pipelines.
    5. Flexible data storage: MongoDB supports multiple storage engines, allowing you to choose the best option for your workload.
    6. Automatic sharding: MongoDB's automatic sharding feature allows you to distribute data across multiple nodes without manual intervention.
    7. Replication and failover: MongoDB supports automatic replication and failover, ensuring high availability and data durability.
    8. ACID transactions: MongoDB now supports multi-document ACID transactions. 
    9. Cloud-based deployment: MongoDB offers cloud-based deployment options, including its Atlas cloud service.
```

```python
Q4.Using the database and the collection,write a code to insert one record,
and insert many records. Use the find() and find_one() methods to print the inserted record.
Ans:-

import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Select the database and collection
db = client["mydatabase"]
collection = db["customers"]

# Insert one record
document = { "name": "Adarsh", "address": "Siwan" }
insert_one_result = collection.insert_one(document)
print(insert_one_result.inserted_id)

# Insert many records
documents = [
  { "name": "suraj", "address": "bihar" },
  { "name": "shivam", "address": "gujrat" },
  { "name": "shyam", "address": "j&k" }
]
insert_many_result = collection.insert_many(documents)
print(insert_many_result.inserted_ids)

# Print the inserted records
print(collection.find_one())
for document in collection.find():
    print(document)
```

```python
Q5.Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.
Ans:-
    The find() method is used to query a MongoDB database and returns a cursor object that you can iterate over to retrieve
    the matching documents. You can pass a filter object to the find() method to specify the selection criteria for the query.


import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Select the database and collection
db = client["mydatabase"]
collection = db["customers"]

# Query the collection
query = { "address": "Park Lane 38" }
results = collection.find(query)

# Print the results
for result in results:
    print(result)
```

```python
Q6.Explain the sort() method. Give an example to demonstrate sorting in MongoDB.
Ans:- The sort() method in MongoDB is used to sort the results of a query in ascending or descending order based 
      on one or more fields.The sort() method takes a dictionary object as an argument, where the keys represent the 
    fields to be sorted on and the values represent the sort order (1 for ascending order and -1 for descending order).
    
import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Select the database and collection
db = client["mydatabase"]
collection = db["customers"]

# Sort the results in ascending order by the "name" field
results = collection.find().sort("name", 1)

# Print the sorted results
for result in results:
    print(result)
```

```python
Q7. Explain why delete_one(), delete_many(), and drop() is used.
Ans:-  
      delete_one() method is used to delete a single document that matches the specified filter from the MongoDB collection.
      If multiple documents match the filter, only the first document encountered is deleted.

    delete_many() method is used to delete all documents that match the specified filter from the MongoDB collection.

    drop() method is used to remove an entire collection or a database from the MongoDB server.

    These methods are used to remove data from a MongoDB collection or database. delete_one() and delete_many() methods
    are useful when you want to selectively remove documents that match a specific criteria, while drop() method is useful 
    when you want to remove an entire collection or database.
```
